From the above data, We can see that there are many columns which has null values

There is no need to remove null values of all the columns. instead we are replacing the null values of killed and wounded column with zero

```python
# fetch the null value of killed column
new_df[new_df['Killed'].isnull()].head(2)
```

```python
# fetch the null value of Wounded column
new_df[new_df['Wounded'].isnull()].head(2)
```

```python
# replace all the null value of killed column with 0
new_df.loc[new_df['Killed'].isna(),'Killed']= 0
```

```python
# replace all the null value of Wounded column with 0
new_df.loc[new_df['Wounded'].isna(),'Wounded']= 0
```

```python
# checking the replaced value
new_df.loc[3:4,'Killed':'Wounded']
```

Successfully replace the null value of killed and wounded column with zero

```python
# now lets create one column named Casualties by using killed and wounded column
```

```python
new_df['Casualties']=new_df['Killed']+new_df['Wounded']
```

##Analysis and visualization of data

###Number of Terrorist Attacks per year/ Yearly growth of terrorist attacks.


```python
# number of attacks per year.
Attacks_per_year=new_df['Year'].value_counts()
Attacks_per_year
```

```python
plt.figure(figsize = (18,6))
sns.lineplot(Attacks_per_year.index, Attacks_per_year.values)
plt.title('Number of Terrorist Attacks per year',fontsize=20)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of attacks',fontsize=15)
plt.show()
```
